```python
from tkinter import *
win = Tk()

counter = 0 
recCounter = 0

def addRec():
    global recCounter
    f = open('db.csv', "a")
    f.write(name.get() + '\t' + mobile.get() + '\t' + address.get() + '\n')
    recCounter = recCounter + 1
    f.close()

def delRec():
    return()

def searchRec():
    return()

def editRec():
    return()

def firstRec():
    f = open('db.csv')
    firstline = f.readline()
    l = firstline.split("\t")
    name.set(l[0])
    mobile.set(l[1])
    address.set(l[2])
    counter = 1
    f.close()

def prevRec():
    return()

def nextRec():
    global counter
    if(counter == 0):
        counter = 1
    f = open("db.csv")
    for i in range(counter - 1):
        line = f.readline()
    line = f.readline()
    l = line.split("\t")
    name.set(l[0])
    mobile.set(l[1])
    address.set(l[2])
    counter = counter + 1
    f.close()

def lastRec():
    return()

label_n = Label(win, text = "Name :")
label_n.grid(row = 1, column = 0)
name = StringVar()
entry_n = Entry(win, textvariable = name)
entry_n.grid(row = 1, column = 2, columnspan = 4)

label_m = Label(win, text = "Mobile Number :")
label_m.grid(row = 2, column = 0)
mobile = StringVar()
entry_m = Entry(win, textvariable = mobile)
entry_m.grid(row = 2, column = 2, columnspan = 4)

label_a = Label(win, text = "Address :")
label_a.grid(row = 3, column = 0)
address = StringVar()
entry_a = Entry(win, textvariable = address)
entry_a.grid(row = 3, column = 2, columnspan = 4)

first_rec = Button(win, command = firstRec, text = "|<")
first_rec.grid(row = 5, column = 0)

prev_rec = Button(win, command = prevRec, text="<")
prev_rec.grid(row = 5, column = 1)

next_rec = Button(win, command = nextRec, text = ">")
next_rec.grid(row = 5, column = 2)

last_rec = Button(win, command = lastRec, text = ">|")
last_rec.grid(row = 5, column = 3)

addrec = Button(win, command = addRec, text = "ADD")
addrec.grid(row = 6, column = 0)

delrec = Button(win, command = delRec, text = "DELETE")
delrec.grid(row = 6, column = 1)

Searchrec = Button(win, command = searchRec, text = "SEARCH")
Searchrec.grid(row = 6, column = 2)

editrec = Button(win, command = editRec, text = "EDIT")
editrec.grid(row = 6, column = 3)

win.mainloop()
```
